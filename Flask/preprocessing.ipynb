```python
import re 
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
stopwords = pd.read_csv('stopwords.txt', header=None)
stopwords.columns = ['word']
stopwords_list = list(stopwords['word'])
# stopwords_list
```

```python
def clean_text(text):    
    text = text.lower()
    text = re.sub(r'[\?,!.*]*', '', text) # replacement of what in the square bracket
    text = re.sub('\w*\d\w*', '', text)# remove digits
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub(re.compile(r'\W*\b\w{1,1}\b'), "", text) #remove short words
    text = re.sub('\n', '', text)
    text = text.split()
    text = [word for word in text if not word in stopwords_list] # remove stopwords
    text = ' '.join(text)
    return text

round1 = lambda x: clean_text(x)
```

```python
data_clean = pd.DataFrame(df.Opinion.apply(round1))
data_clean
```

```python
data_clean['Sentiment'] = df['Sentiment']
data_clean
```

```python
# df.to_pickle('corpus_cleaned.pkl')
```

```python
df_x =data_clean['Opinion']
df_y = data_clean['Sentiment']
```

```python
# Document-Term Matrix - word counts in matrix format
cv = CountVectorizer()
data_cv = cv.fit_transform(data_clean.Opinion)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm.index = data_clean.index
data_dtm
```

```python
# Let's pickle it for later use
data_dtm.to_pickle("dtm.pkl")
```

```python
# pickle the cleaned data and the CountVectorizer object
data_clean.to_pickle('data_clean.pkl')
pickle.dump(cv, open("cv.pkl", "wb"))
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size = 0.25, random_state = 4)
```

```python
x_train_countvectorizer = cv.fit_transform(x_train)
```

```python
# print(x_train[0])
# print(x_train_countvectorizer[0])
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(x_train_countvectorizer, y_train)
```

```python
x_test_countvectorizer = cv.transform(x_test)
```

```python
pred = mnb.predict(x_test_countvectorizer)
# print(pred)
```

```python
rezult = np.array(y_test)
# print(rezult)
```

```python
count = 0
for i in range(len(pred)):
    if pred[i] == rezult[i]:
        count += 1    

print(count)
print(len(pred))
```

```python
accuracy = count/len(pred)
print(accuracy)
```
